## Obtendo os 8 datasets

```python
smart_tv_biggest_median_upload_df = smart_tv_df[smart_tv_df["hour"]==smart_tv_df.groupby("hour")["log_bytes_up"].median().idxmax()]
smart_tv_biggest_avg_upload_df = smart_tv_df[smart_tv_df["hour"]==smart_tv_df.groupby("hour")["log_bytes_up"].mean().idxmax()] 

smart_tv_biggest_median_download_df = smart_tv_df[smart_tv_df["hour"]==smart_tv_df.groupby("hour")["log_bytes_down"].median().idxmax()]
smart_tv_biggest_avg_download_df = smart_tv_df[smart_tv_df["hour"]==smart_tv_df.groupby("hour")["log_bytes_down"].mean().idxmax()] 

chromecast_biggest_median_upload_df = chromecast_df[chromecast_df["hour"]==chromecast_df.groupby("hour")["log_bytes_up"].median().idxmax()]
chromecast_biggest_avg_upload_df = chromecast_df[chromecast_df["hour"]==chromecast_df.groupby("hour")["log_bytes_up"].mean().idxmax()] 

chromecast_biggest_median_download_df = chromecast_df[chromecast_df["hour"]==chromecast_df.groupby("hour")["log_bytes_down"].median().idxmax()]
chromecast_biggest_avg_download_df = chromecast_df[chromecast_df["hour"]==chromecast_df.groupby("hour")["log_bytes_down"].mean().idxmax()] 
```

```python
datasets = [smart_tv_biggest_median_upload_df, smart_tv_biggest_avg_upload_df, smart_tv_biggest_median_download_df, smart_tv_biggest_avg_download_df, 
            chromecast_biggest_median_upload_df, chromecast_biggest_avg_upload_df, chromecast_biggest_median_download_df, chromecast_biggest_avg_download_df]
```

# MLE

```python
from scipy.stats import norm, gamma
from scipy.stats import probplot, pearsonr, power_divergence
```

## Gamma

### Bytes Up

```python
bytes_up_gamma_fns = []
for ix, dataset in enumerate(datasets):
  alfa, loc, scale = gamma.fit(dataset["log_bytes_up"])
  print(f"Gamma fit de Bytes Up do dataset {ix}. Alfa: {alfa:.2f}; Loc: {loc:.2f}; Beta: {1/scale:.2f}") 
  bytes_up_gamma_fns.append((alfa,loc,scale))
```

### Bytes Down

```python
bytes_down_gamma_fns = []
for ix, dataset in enumerate(datasets):
  alfa, loc, scale = gamma.fit(dataset["log_bytes_down"])
  print(f"Gamma fit de Bytes Down do dataset {ix}. Alfa: {alfa:.2f}; Loc: {loc:.2f}; 1/Beta: {1/scale:.2f}") 
  bytes_down_gamma_fns.append((alfa,loc,scale))
```

## Guassiana

### Bytes Up

```python
bytes_up_gaussian_fns = []
for ix, dataset in enumerate(datasets):
  mean,std = norm.fit(dataset["log_bytes_up"])
  print(f"Gaussian fit de Bytes Up do dataset {ix}. Média: {mean:.2f}; Desvio padrão: {std:.2f}") 
  bytes_up_gaussian_fns.append((mean,std))
```

### Bytes Down

```python
bytes_down_gaussian_fns = []
for ix, dataset in enumerate(datasets):
  mean,std = norm.fit(dataset["log_bytes_down"])
  print(f"Gaussian fit de Bytes Down do dataset {ix}. Média: {mean:.2f}; Desvio padrão: {std:.2f}") 
  bytes_down_gaussian_fns.append((mean,std))
```

# Análise correlação

```python
correlations_tests = [(1,3),(2,4),(5,7),(6,8)]
```

```python
for c in correlations_tests:
  dataset_1 = datasets[c[0]-1].sort_values(by=['hour'])["log_bytes_up"]
  dataset_2 = datasets[c[1]-1].sort_values(by=['hour'])["log_bytes_down"] 

  if dataset_1.shape[0] > dataset_2.shape[0]: dataset_1 = dataset_1[:dataset_2.shape[0]]
  else: dataset_2 = dataset_2[:dataset_1.shape[0]]

  pearson, p = pearsonr(dataset_1, dataset_2)
  print(f"Correlação entre datasets {c[0]} e {c[1]}. Pearson: {pearson}; P-valor: {p}")
```

```python
for c in correlations_tests:
  fig = plt.figure() 
  dataset_1 = datasets[c[0]-1].sort_values(by=['hour'])["log_bytes_up"]
  dataset_2 = datasets[c[1]-1].sort_values(by=['hour'])["log_bytes_down"] 
  
  if dataset_1.shape[0] > dataset_2.shape[0]: dataset_1 = dataset_1[:dataset_2.shape[0]]
  else: dataset_2 = dataset_2[:dataset_1.shape[0]]

  plt.scatter(dataset_1, dataset_2)
  plt.xlabel(f"Bytes Up Dataset {c[0]}")
  plt.ylabel(f"Bytes Up Dataset {c[1]}")
  plt.title(f"Scatter Plot entre datasets {c[0]} e {c[1]}")
```

# G-Testes

```python
g_tests = [(1,5),(2,6),(3,7),(4,8)]
```

```python
for d1,d2 in g_tests:
  bins = sturges(datasets[d1-1]["log_bytes_up"])
  hist_1,_ = np.histogram(datasets[d1-1]["log_bytes_up"], bins=bins)
  hist_2,_ = np.histogram(datasets[d2-1]["log_bytes_down"], bins=bins)

  hist_1 = hist_1/hist_1.sum()
  hist_2 = hist_2/hist_2.sum()
  _, p_value = power_divergence(hist_1, hist_2, lambda_='log-likelihood') 
  print(f"P-valor {p_value} do G-Test entre datasets {d1} e {d2}")
```
